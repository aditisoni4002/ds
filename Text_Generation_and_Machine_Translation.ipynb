Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

Answer:

Probabilistic modeling plays a fundamental and central role in generative models. At its core, a generative model aims to learn the underlying probability distribution of the training data. If it successfully learns this distribution, it can then sample from it to generate new data points that resemble the original training data.

Specifically, generative models often focus on learning the joint probability distribution $P(x, y)$$P(x, y)$ of inputs $x$$x$ and labels $y$$y$ (or just $P(x)$$P(x)$ if there are no labels, as in unsupervised generation). By understanding how inputs and labels occur together, the model can generate new $x$$x$ values (e.g., images, text) that fit the learned patterns.

Key aspects of probabilistic modeling in generative models:

Learning Data Distribution: The primary goal is to model $P(data)$$P(data)$ or $P(x)$$P(x)$. This can be achieved through various techniques, such as explicit density estimation (e.g., PixelCNN, some VAEs) or implicit density estimation (e.g., GANs, which learn to sample from the distribution without explicitly defining its density function).
Sampling: Once the probability distribution is learned (or approximated), the model can sample new data points from it. This sampling process is what leads to the generation of novel content.
Latent Space Representation: Many generative models (like VAEs or GANs) utilize a latent space, often assumed to follow a simple prior distribution (e.g., a Gaussian distribution). The model learns a mapping from this latent space to the complex data distribution. This latent space captures meaningful, disentangled features of the data in a probabilistic manner.
Inference (for some models): In models like VAEs, there's also an inference network that learns to map data back into the latent space, providing a probabilistic representation of an input data point within that space ($P(z|x)$$P(z|x)$).
Generative Models vs. Discriminative Models:

The fundamental difference lies in what they aim to model and what tasks they are designed for:

What they Model:

Generative Models: Learn the joint probability distribution $P(x, y)$$P(x, y)$ (or $P(x)$$P(x)$). They understand how the data is generated. Because they learn $P(x, y)$$P(x, y)$, they can also derive $P(y|x)$$P(y|x)$ (for classification) and $P(x|y)$$P(x|y)$ (for generating inputs given a class).
Discriminative Models: Learn the conditional probability distribution $P(y|x)$$P(y|x)$. They focus on distinguishing between different classes or predicting a target given an input. They do not model the underlying data generation process.
Primary Task:

Generative Models: Primarily designed for generating new data. They can also perform classification by calculating $P(y|x)$$P(y|x)$ from $P(x,y)$$P(x,y)$. Examples: GANs, VAEs, Autoregressive models (e.g., GPT for text generation).
Discriminative Models: Primarily designed for classification (e.g., assigning a label to an image) or regression (e.g., predicting a numerical value). They cannot directly generate new data. Examples: Logistic Regression, Support Vector Machines (SVMs), standard Feedforward Neural Networks for classification.
Nature of Learning:

Generative Models: More complex to train, as they learn the entire data distribution. They often require more data and computational resources. Can be more robust to missing data and detect anomalies more easily.
Discriminative Models: Generally simpler to train as they only focus on the decision boundary. They are often more accurate for classification tasks when sufficient data is available, as they don't waste capacity modeling aspects of the data that are irrelevant to discrimination.


Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

Answer:

Both Autoencoders (AEs) and Variational Autoencoders (VAEs) are neural network architectures used for unsupervised learning, primarily for dimensionality reduction and learning latent representations of data. While they share a common encoder-decoder structure, their underlying probabilistic foundations and therefore their utility, especially in text generation, differ significantly.

Autoencoders (AEs)
An Autoencoder is designed to learn an efficient, compressed representation (encoding) of input data. It consists of two main parts:

Encoder: Maps the input data $X$$X$ to a lower-dimensional latent space representation $Z$$Z$, such that $Z = f(X)$$Z = f(X)$.
Decoder: Reconstructs the input data $X'$$X'$ from the latent representation $Z$$Z$, such that $X' = g(Z)$$X' = g(Z)$.
The goal of an AE is to minimize the reconstruction error between the input $X$$X$ and its reconstruction $X'$$X'$.

In the context of text generation:

Latent Space: The latent space learned by a standard AE is typically a point estimate. There's no explicit structure imposed on this latent space to ensure continuity or completeness.
Generation Limitations: While an AE can learn representations useful for tasks like text denoising or summarization, it is generally not well-suited for generating new, coherent text. If you sample a random point from its latent space and feed it to the decoder, there's no guarantee that the point will correspond to a meaningful or grammatically correct text. The latent space can have 'holes' or areas that don't correspond to any valid data.
Text Reconstruction: AEs can be effective for reconstructing input text, learning salient features for representation, but their generative capability for novel text is limited due to the unstructured nature of their latent space.
Variational Autoencoders (VAEs)
A Variational Autoencoder addresses the generative limitations of standard AEs by introducing a probabilistic approach. Instead of learning a direct mapping to a latent vector, VAEs learn the parameters of a probability distribution in the latent space.

Encoder: Maps the input data $X$$X$ to parameters (mean $\mu$$\mu$ and variance $\sigma$$\sigma$) of a probability distribution (typically a Gaussian) in the latent space. So, for each input $X$$X$, the encoder outputs $P(Z|X)$$P(Z|X)$.
Reparameterization Trick: To allow backpropagation through this probabilistic step, a sample $z$$z$ is drawn from $P(Z|X)$$P(Z|X)$ using the reparameterization trick: $z = \mu + \sigma \odot \epsilon$$z = \mu + \sigma \odot \epsilon$, where $\epsilon \sim N(0, I)$$\epsilon \sim N(0, I)$.
Decoder: Reconstructs the input data $X'$$X'$ from the sampled latent vector $z$$z$, aiming to maximize the likelihood $P(X|Z)$$P(X|Z)$.
VAEs optimize a loss function that has two components:

Reconstruction Loss: Measures how well the decoder reconstructs the input.
KL Divergence Loss: Forces the latent distribution $P(Z|X)$$P(Z|X)$ to be close to a prior distribution (e.g., a standard normal distribution $N(0, I)$$N(0, I)$). This regularization term encourages the latent space to be continuous and well-structured, preventing the 'holes' found in AE latent spaces.
In the context of text generation:

Probabilistic Latent Space: The primary difference is the probabilistic nature of the latent space. By forcing the latent distributions to conform to a prior, the VAE ensures that the latent space is continuous and traversable. This means that points sampled from the prior distribution are likely to correspond to meaningful data points.
Effective Generation: This structured latent space makes VAEs much more suitable for generating novel text. You can sample a random vector $z$$z$ from the simple prior distribution (e.g., a standard normal distribution) and feed it to the decoder. Because the encoder learned to map diverse inputs to a continuous, well-behaved latent space, the decoder is more likely to generate coherent and grammatically plausible text from these random latent samples.
Latent Space Interpolation: The continuous latent space also allows for meaningful interpolation. Moving smoothly between two latent vectors can result in a smooth transition between the generated texts, which is useful for tasks like style transfer or controlling aspects of text generation.
Text Reconstruction: VAEs also perform text reconstruction, similar to AEs, but with the added benefit of a disentangled and meaningful latent representation.
Summary for Text Generation:

Feature	Autoencoder (AE)	Variational Autoencoder (VAE)
Latent Space	Deterministic point estimate; can have 'holes'.	Probabilistic distribution (e.g., Gaussian); continuous & structured.
Generation	Poor; random samples from latent space often produce garbage.	Good; random samples from prior produce meaningful outputs.
Loss Function	Reconstruction loss.	Reconstruction loss + KL Divergence loss (regularization).
Primary Use	Dimensionality reduction, data compression, feature learning.	Generative modeling, continuous latent representations, data generation.
Text Generation	Limited to reconstruction; not for novel text synthesis.	Capable of synthesizing novel, coherent text by sampling from a learned continuous latent space.
In essence, VAEs add a layer of probabilistic robustness to Autoencoders, making their latent spaces truly useful for generating new and diverse data, including text, by ensuring continuity and meaningfulness across the latent representation.

Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?

Answer:

Working of Attention Mechanisms in Neural Machine Translation (NMT)
Neural Machine Translation (NMT) models typically employ an encoder-decoder architecture. The encoder processes the source sentence (e.g., English) into a sequence of contextualized representations (hidden states), and the decoder then generates the target sentence (e.g., French) one word at a time based on these representations.

Before attention, decoders would rely solely on a single fixed-size context vector generated by the encoder, which was a bottleneck for long sentences as it struggled to capture all relevant information. Attention mechanisms address this by allowing the decoder to 'attend' to different parts of the source sentence at each step of generating the target sentence.

Here's how it generally works:

Encoder Output: The encoder processes the source sequence $(x_1, x_2, ..., x_N)$$(x_1, x_2, ..., x_N)$ and produces a sequence of hidden states $(h_1, h_2, ..., h_N)$$(h_1, h_2, ..., h_N)$, where each $h_i$$h_i$ represents the context around source word $x_i$$x_i$.

Decoder State at Current Step: At each decoding step $t$$t$, the decoder generates a new target word $y_t$$y_t$. To do this, it has its own current hidden state, $s_t$$s_t$, which summarizes what has been generated so far.

Alignment Score Calculation: An 'alignment score' (or 'energy score') is calculated between the current decoder state $s_t$$s_t$ and each of the encoder's hidden states $h_i$$h_i$. This score measures how relevant each source word's representation $h_i$$h_i$ is to generating the current target word $y_t$$y_t$. Common scoring functions include dot product, concatenation with a feed-forward network, or an additive (Bahdanau) attention.

$e_{ti} = ext{score}(s_t, h_i)$$e_{ti} = ext{score}(s_t, h_i)$
Softmax Normalization: The alignment scores are then passed through a softmax function to get 'attention weights' ($\alpha_{ti}$$\alpha_{ti}$). These weights are normalized probabilities, indicating how much attention the decoder should pay to each source word $x_i$$x_i$ when generating $y_t$$y_t$.

$\alpha_{ti} = \frac{\exp(e_{ti})}{\sum_{k=1}^N \exp(e_{tk})}$$\alpha_{ti} = \frac{\exp(e_{ti})}{\sum_{k=1}^N \exp(e_{tk})}$
Context Vector Creation: A weighted sum of the encoder's hidden states is computed using these attention weights. This creates a 'context vector' ($c_t$$c_t$) that is dynamically tailored for generating the current target word $y_t$$y_t$, focusing on the most relevant parts of the source sentence.

$c_t = \sum_{i=1}^N \alpha_{ti} h_i$$c_t = \sum_{i=1}^N \alpha_{ti} h_i$
Decoder Prediction: The context vector $c_t$$c_t$ is then concatenated with the decoder's current hidden state $s_t$$s_t$ and fed into the next layer of the decoder (e.g., a feed-forward network and softmax) to predict the next target word $y_t$$y_t$.

This process repeats for every word generated in the target sequence.

Why are Attention Mechanisms Critical?
Attention mechanisms are critical in NMT for several reasons:

Handling Long Sentences (Overcoming Bottleneck): Without attention, the encoder has to compress all information of the source sentence into a single, fixed-size context vector. For long sentences, this leads to significant information loss. Attention allows the decoder to selectively access relevant information from the entire source sequence at each step, effectively bypassing this bottleneck.

Improved Performance and Accuracy: By providing the decoder with a dynamic and relevant context, attention mechanisms significantly improve the quality of translations, especially for complex and longer sentences. This leads to more fluent, grammatically correct, and semantically accurate outputs.

Interpretability (Alignment Visualization): The attention weights ($\alpha_{ti}$$\alpha_{ti}$) can be visualized as an alignment matrix, showing which source words the model is focusing on when generating each target word. This provides a valuable degree of interpretability, allowing researchers and practitioners to understand why the model makes certain translation choices.

Handling Word Reordering: Different languages have different word orders. Attention allows the model to flexibly align source and target words, even if their positions are significantly different, without being constrained by the sequential order of the encoder's final hidden state.

Focusing on Salient Information: At any given step, the model can 'concentrate' its computational resources on the most pertinent parts of the input, rather than processing the entire input uniformly, making the process more efficient and effective.

In essence, attention mechanisms transformed NMT by enabling models to cope with the complexities of language translation, leading to the highly performant and widely adopted NMT systems we see today.

Ethical Guidelines and Regulation:

Need for Standards: There's a growing need for industry standards, ethical guidelines, and potentially regulations for the development and use of generative AI in creative fields to mitigate these risks.
Transparency: Models should be developed with a degree of transparency regarding their training data, limitations, and potential biases.
Addressing these ethical considerations requires a multidisciplinary approach involving AI developers, ethicists, legal experts, artists, and policymakers to ensure that generative AI enhances rather than diminishes human creativity and adheres to societal values.

#### VAE Model Definition

Here we define the VAE architecture, including the encoder, the reparameterization trick, and the decoder. The encoder maps input sequences to a latent distribution, from which we sample a latent vector. The decoder then uses this latent vector to reconstruct the input sequence. A crucial part is the VAE loss, which combines reconstruction loss with a KL divergence term to encourage a well-structured latent space.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Lambda, Reshape, TimeDistributed
from tensorflow.keras import backend as K
from tensorflow.keras.losses import sparse_categorical_crossentropy
import numpy as np
# Import initializers explicitly if needed, but trying string names first

# VAE model parameters
embed_dim = 100
lstm_units = 128
latent_dim = 64

# 2. Build a basic VAE model for text reconstruction

# Encoder
encoder_inputs = Input(shape=(max_sequence_len,))
encoder_embedding = Embedding(vocabulary_size, embed_dim, mask_zero=True)(encoder_inputs)
encoder_lstm = LSTM(lstm_units, return_state=True, kernel_initializer='random_normal', recurrent_initializer='random_normal')(encoder_embedding)
_, state_h, state_c = encoder_lstm # We only need the states
encoder_states = [state_h, state_c]

# Latent space for VAE - mean and log variance
z_mean = Dense(latent_dim)(state_h)
z_log_var = Dense(latent_dim)(state_h)

# Reparameterization trick
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon

z = Lambda(sampling, output_shape=(latent_dim,))([z_mean, z_log_var])

# Decoder
decoder_inputs = Input(shape=(max_sequence_len,))
decoder_embedding = Embedding(vocabulary_size, embed_dim, mask_zero=True)(decoder_inputs)
decoder_lstm = LSTM(lstm_units, return_sequences=True, kernel_initializer='random_normal', recurrent_initializer='random_normal')

# The decoder LSTM needs to receive the latent vector 'z' as its initial state
# To achieve this, we can project 'z' to the shape of the LSTM states (h, c)
decoder_state_h = Dense(lstm_units)(z)
decoder_state_c = Dense(lstm_units)(z)
decoder_states = [decoder_state_h, decoder_state_c]

decoder_outputs = decoder_lstm(decoder_embedding, initial_state=decoder_states)
decoder_dense = TimeDistributed(Dense(vocabulary_size, activation='softmax'))(decoder_outputs)

# Define VAE model
vae = Model([encoder_inputs, decoder_inputs], decoder_dense)

# VAE Loss = Reconstruction Loss + KL-Divergence Loss
# Reconstruction Loss (Sparse Categorical Crossentropy for sequence prediction)
def vae_reconstruction_loss(y_true, y_pred):
    # y_true needs to be one-hot encoded or similar for TimeDistributed(Dense(softmax))
    # However, for sparse_categorical_crossentropy, y_true should be integer labels.
    # K.flatten(y_true) flattens to (batch*timesteps,) and K.flatten(y_pred) to (batch*timesteps, vocab_size)
    # The mask_zero in embedding layer implicitly handles padding, ensuring padded positions don't contribute to loss.
    reconstruction_loss = K.sum(sparse_categorical_crossentropy(y_true, y_pred))
    return reconstruction_loss

def vae_kl_loss(z_mean, z_log_var):
    kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
    return kl_loss

def vae_loss(y_true, y_pred, z_mean, z_log_var):
    recon_loss = vae_reconstruction_loss(y_true, y_pred)
    kl_loss = vae_kl_loss(z_mean, z_log_var)
    return recon_loss + kl_loss

# Custom VAE loss function for compilation
def custom_vae_loss(y_true, y_pred):
    return vae_loss(y_true, y_pred, z_mean, z_log_var)

vae.compile(optimizer='adam', loss=custom_vae_loss)

vae.summary()
```

#### Training the VAE

Now, we train the VAE model using the preprocessed data. The input to the VAE will be the padded sequences for both the encoder and decoder. The target for the loss function is also the padded sequences, as the VAE aims to reconstruct its input.

```python
# 3. Train the model

# The input to the VAE is two inputs: encoder_inputs and decoder_inputs.
# For reconstruction, both will be X_train (padded_sequences).
# The target is y_train (also padded_sequences).
vae.fit([X_train, X_train], y_train, epochs=100, batch_size=1, verbose=0)
print("VAE model trained successfully!")
```

```python
# Build encoder model (to get z_mean from input)
encoder_model = Model(encoder_inputs, [z_mean, z_log_var])

# Build decoder model (to generate sequence from z)
latent_input = Input(shape=(latent_dim,))

# Project latent input to LSTM states
decoder_state_h_decode = Dense(lstm_units)(latent_input)
decoder_state_c_decode = Dense(lstm_units)(latent_input)
decoder_states_decode = [decoder_state_h_decode, decoder_state_c_decode]

# The decoder LSTM itself needs to be reusable
decoder_lstm_inference = LSTM(lstm_units, return_sequences=True, return_state=True, kernel_initializer='glorot_uniform') # Add initializer here too

# We'll use a placeholder for decoder_input_inference for predicting word by word
decoder_input_inference = Input(shape=(1,))
decoder_embedding_inference = Embedding(vocabulary_size, embed_dim, mask_zero=True)(decoder_input_inference)

decoder_outputs_inf, state_h_inf, state_c_inf = decoder_lstm_inference(
decoder_embedding_inference, initial_state=decoder_states_decode
)

decoder_dense_inference = TimeDistributed(Dense(vocabulary_size, activation='softmax'))(decoder_outputs_inf)

decoder_model = Model(
    [decoder_input_inference, latent_input, decoder_state_h_decode, decoder_state_c_decode],
    [decoder_dense_inference, state_h_inf, state_c_inf]
)


# Helper function to convert sequence of word indices back to text
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

def decode_sequence(input_seq, max_len, tokenizer_obj, encoder_m, decoder_m, latent_dim_val):
    # Get latent vector from encoder
    z_mean, _ = encoder_m.predict(input_seq, verbose=0)
    z = z_mean # Use the mean for deterministic decoding

    # State value to carry over between each step of word generation
    # Need to get layer by name as the indices might change after recompilation/rebuilding
    # Using get_layer('dense_2') and get_layer('dense_3') which are for decoder_state_h and decoder_state_c projections from z
    states_value = [Model(encoder_inputs, z_mean).predict(input_seq), Model(encoder_inputs, z_log_var).predict(input_seq)] # this part was problematic, re-evaluate how to get states from z
    # Correct way to get initial states for inference decoder from latent vector
    h_input = Input(shape=(latent_dim,))
    c_input = Input(shape=(latent_dim,))
    decoder_h_layer = vae.get_layer('dense_2')
    decoder_c_layer = vae.get_layer('dense_3')
    get_h_from_z = Model(h_input, decoder_h_layer(h_input))
    get_c_from_z = Model(c_input, decoder_c_layer(c_input))

    initial_decoder_h = get_h_from_z.predict(z, verbose=0)
    initial_decoder_c = get_c_from_z.predict(z, verbose=0)
    states_value = [initial_decoder_h, initial_decoder_c]

    # Seed input for the decoder (e.g., a dummy token to start generation)
    current_word_idx = np.array([[0]]) # Dummy start token (padding index)

    decoded_sentence = []
    for _ in range(max_len):
        output_tokens, h, c = decoder_m.predict([current_word_idx, z, states_value[0], states_value[1]], verbose=0)

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])

        # If it's a padding token or unk, stop or skip
        if sampled_token_index == 0 or (tokenizer_obj.num_words is not None and sampled_token_index > tokenizer_obj.num_words):
            break # Stop if padding or OOV

        sampled_word = tokenizer_obj.index_word.get(sampled_token_index, '')
        if sampled_word == '<unk>' or sampled_word == '': # Stop if it predicts OOV or empty string from get
            break

        decoded_sentence.append(sampled_word)

        # Update the target sequence (of length 1).
        current_word_idx = np.array([[sampled_token_index]])
        # Update states
        states_value = [h, c]

    return ' '.join(decoded_sentence)


def generate_sentence_from_latent(decoder_m, latent_vec, max_len, tokenizer_obj):
    # Re-obtain the dense layers from the main VAE to project latent_vec to LSTM states
    decoder_h_layer = vae.get_layer('dense_2')
    decoder_c_layer = vae.get_layer('dense_3')

    h_input = Input(shape=(latent_dim,))
    c_input = Input(shape=(latent_dim,))
    get_h_from_z = Model(h_input, decoder_h_layer(h_input))
    get_c_from_z = Model(c_input, decoder_c_layer(c_input))

    initial_decoder_h = get_h_from_z.predict(latent_vec, verbose=0)
    initial_decoder_c = get_c_from_z.predict(latent_vec, verbose=0)
    states_value = [initial_decoder_h, initial_decoder_c]

    current_word_idx = np.array([[0]]) # Start with a dummy token (padding index)

    decoded_sentence = []
    for _ in range(max_len):
        output_tokens, h, c = decoder_m.predict([current_word_idx, latent_vec, states_value[0], states_value[1]], verbose=0)

        sampled_token_index = np.argmax(output_tokens[0, -1, :])

        if sampled_token_index == 0 or (tokenizer_obj.num_words is not None and sampled_token_index > tokenizer_obj.num_words):
            break

        sampled_word = tokenizer_obj.index_word.get(sampled_token_index, '')
        if sampled_word == '<unk>' or sampled_word == '':
            break

        decoded_sentence.append(sampled_word)

        current_word_idx = np.array([[sampled_token_index]])
        states_value = [h, c]

    return ' '.join(decoded_sentence)


# --- Demonstration ---

print("\n--- Text Reconstruction ---")
for i, input_seq in enumerate(X_train):
    original_text = data[i]
    # Ensure input_seq is reshaped correctly for the encoder_model.predict
    reconstructed_text = decode_sequence(np.array([input_seq]), max_sequence_len, tokenizer, encoder_model, decoder_model, latent_dim)
    print(f"Original: '{original_text}'")
    print(f"Reconstructed: '{reconstructed_text}'\n")

print("\n--- Text Generation from Latent Space ---")
num_generations = 3
for _ in range(num_generations):
    # Sample random latent vector from a normal distribution
    random_latent_vec = np.random.normal(size=(1, latent_dim))
    generated_text = generate_sentence_from_latent(decoder_model, random_latent_vec, max_sequence_len, tokenizer)
    print(f"Generated: '{generated_text}'")
```

### Question 7: Using a pre-trained GPT model for language translation

**Task:** Explain how to use a pre-trained GPT model (or any suitable transformer-based model) for language translation. Provide a code example demonstrating how to translate English to French.

**Answer:**

While general GPT models like those from OpenAI can be fine-tuned for translation, models specifically trained for sequence-to-sequence tasks, such as those often found in the Hugging Face `transformers` library, are typically more efficient and performant for direct machine translation. For English to French translation, a common choice is the `Helsinki-NLP/opus-mt-en-fr` model, which is a transformer-based model optimized for this task.

The process generally involves:
1.  **Installation**: Install the `transformers` library.
2.  **Model and Tokenizer Loading**: Load a pre-trained translation model and its corresponding tokenizer.
3.  **Tokenization**: Convert the source English text into numerical tokens that the model understands.
4.  **Translation Generation**: Feed the tokenized input to the model to generate translated output tokens.
5.  **Decoding**: Convert the output tokens back into human-readable French text.

```python
# Install the transformers library if not already installed
!pip install transformers
```

#### Translation Code Example

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# 1. Load the pre-trained tokenizer and model
# We'll use a model from Helsinki-NLP for English to French translation
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

print(f"Loaded model: {model_name}")

# 2. Define the English sentence to translate
english_sentence = "The quick brown fox jumps over the lazy dog."
print(f"\nOriginal English sentence: '{english_sentence}'")

# 3. Tokenize the input sentence
# The return_tensors='pt' argument specifies to return PyTorch tensors, but Transformers can also work with TensorFlow.
# For this example, we'll keep it simple and let it use the default framework detected or convert manually if needed.
input_ids = tokenizer(english_sentence, return_tensors="pt").input_ids

# 4. Generate the translation
# The generate method handles the entire decoding process.
# max_length can be adjusted based on expected translation length.
# num_beams for beam search, typically improves translation quality.
output_ids = model.generate(input_ids, max_length=50, num_beams=5, early_stopping=True)

# 5. Decode the output tokens back to French text
french_translation = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print(f"French translation: '{french_translation}'")
```

**Answer:**

We will implement this using TensorFlow. The process involves:
1.  **Data Preparation**: Create a small dataset, preprocess it, tokenize, build vocabulary, and pad sequences.
2.  **Encoder**: Process the input sequence into a context vector.
3.  **Attention Mechanism**: Allow the decoder to focus on relevant parts of the encoder's output.

```python
# Install necessary libraries
!pip install tensorflow numpy matplotlib
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input, Embedding, GRU, Dense, Attention, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import unicodedata
import re
```

#### Data Preparation

We will create a small, synthetic dataset of English-Spanish sentence pairs. We'll preprocess the text by normalizing Unicode characters, removing special characters, and adding start (`<start>`) and end (`<end>`) tokens. Then, we'll tokenize the text and pad sequences to a uniform length.

```python
# 1. Data Preparation

# Small dataset (English-Spanish)
data = [
    ("Go.", "Ve."),
    ("Hi.", "Hola."),
    ("Run!", "¡Corre!"),
    ("Run.", "Corra."),
    ("Who?", "¿Quién?"),
    ("Wow!", "¡Guau!"),
    ("Fire!", "¡Fuego!"),
    ("Help!", "¡Ayuda!"),
    ("Jump!", "¡Salte!"),
    ("Stop!", "¡Alto!"),
    ("Stop.", "Alto."),
    ("Wait!", "¡Espera!"),
    ("Wait.", "Espera."),
    ("I try.", "Lo intento."),
    ("I won!", "¡Ganamos!"),
    ("I won.", "Ganamos."),
    ("I see.", "Ya veo."),
    ("I see.", "Entiendo."),
    ("I left.", "Me fui."),
    ("I know.", "Lo sé."),
    ("I know.", "Sé."),
    ("I hate.", "Odio.")
]

def unicode_to_ascii(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')

def preprocess_sentence(s):
    s = unicode_to_ascii(s.lower().strip())
    s = re.sub(r"([?.!,¿¡])", r" \1", s)
    s = re.sub(r'["]', '', s)
    s = re.sub(r"[^a-zA-Z?.!,¿¡]+", r" ", s)
    s = s.strip()
    s = '<start> ' + s + ' <end>'
    return s

english_sentences = [preprocess_sentence(pair[0]) for pair in data]
spe_sentences = [preprocess_sentence(pair[1]) for pair in data]

print("Example English sentence: ", english_sentences[0])
print("Example Spanish sentence: ", spe_sentences[0])

def create_tokenizer(lang_sentences):
    tokenizer = Tokenizer(filters='', oov_token='<unk>')
    tokenizer.fit_on_texts(lang_sentences)
    tensor = tokenizer.texts_to_sequences(lang_sentences)
    tensor = pad_sequences(tensor, padding='post')
    return tokenizer, tensor

input_tokenizer, input_tensor = create_tokenizer(english_sentences)
target_tokenizer, target_tensor = create_tokenizer(spe_sentences)

max_input_len = input_tensor.shape[1]
max_target_len = target_tensor.shape[1]

print(f"Max English sentence length: {max_input_len}")
print(f"Max Spanish sentence length: {max_target_len}")
print(f"English Vocabulary Size: {len(input_tokenizer.word_index) + 1}")
print(f"Spanish Vocabulary Size: {len(target_tokenizer.word_index) + 1}")

# Create tf.data.Dataset
BATCH_SIZE = 4
BUFFER_SIZE = len(input_tensor)

dataset = tf.data.Dataset.from_tensor_slices((input_tensor, target_tensor)).shuffle(BUFFER_SIZE)
dataset = dataset.batch(BATCH_SIZE, drop_remainder=True)

example_input_batch, example_target_batch = next(iter(dataset))
print(f"\nExample input batch shape: {example_input_batch.shape}")
print(f"Example target batch shape: {example_target_batch.shape}")
```

**Task:** Use the provided short poetry dataset to simulate poem generation with a pre-trained GPT model:
`["Roses are red, violets are blue,", "Sugar is sweet, and so are you.", "The moon glows bright in silent skies,", "A bird sings where the soft wind sighs."]`
Using this dataset as a reference for poetic structure and language, generate a new 2-4 line poem using a pre-trained GPT model (such as GPT-2). You may simulate fine-tuning by prompting the model with similar poetic patterns.
Include your code, the prompt used, and the generated poem in your answer.

**Answer:**

To simulate poem generation using a pre-trained GPT model like GPT-2, we leverage its ability to continue text based on a given prompt. By crafting a prompt that mimics the structure and tone of the provided poetry dataset, we can guide the model to generate a new poem in a similar style. This acts as a form of 'in-context learning' or 'few-shot learning', where the model picks up on patterns without explicit fine-tuning on the dataset.

The steps are:
1.  **Load a pre-trained GPT model and its tokenizer** (e.g., `gpt2`).
2.  **Define the reference poetry dataset**.
3.  **Construct a prompt** that includes elements from the dataset to set the poetic context.
4.  **Generate text** using the model, controlling parameters like `max_length`, `num_return_sequences`, and `temperature` to influence the output's creativity and length.
5.  **Display the prompt and the generated poem**.

```python
from transformers import pipeline, set_seed

# Set a seed for reproducibility
set_seed(42)

# Load the GPT-2 model for text generation
generator = pipeline('text-generation', model='gpt2')

# Provided poetry dataset
poetry_dataset = [
    "Roses are red, violets are blue,",
    "Sugar is sweet, and so are you.",
    "The moon glows bright in silent skies,",
    "A bird sings where the soft wind sighs."
]

# Construct a prompt that mimics the style and structure of the dataset
# We provide a few examples to guide the model, simulating a 'fine-tuning' effect.
prompt = (
    f"{poetry_dataset[0]}\n"  # Roses are red, violets are blue,
    f"{poetry_dataset[1]}\n"  # Sugar is sweet, and so are you.
    f"{poetry_dataset[2]}\n"  # The moon glows bright in silent skies,
    f"{poetry_dataset[3]}\n"  # A bird sings where the soft wind sighs.
    "Nature's beauty, a gentle art,\n" # Starting a new line for the model to continue
)

print("--- Prompt Used ---")
print(prompt)

# Generate the poem
# num_return_sequences=1 to get one generation
# max_new_tokens to control length
# temperature controls randomness (lower for less random, higher for more creative)
generated_poems = generator(
    prompt,
    max_new_tokens=20, # Generate up to 20 new tokens to keep it around 2-4 lines
    num_return_sequences=1,
    temperature=0.7,
    top_k=50,
    pad_token_id=generator.tokenizer.eos_token_id # Crucial for GPT-2 when generating shorter texts
)

print("\n--- Generated Poem ---")
# Extract and print only the generated part that continues the prompt
generated_text = generated_poems[0]['generated_text'][len(prompt):].strip()

# To ensure it's multi-line and looks like a poem, we can split by newline and take a few lines
poem_lines = generated_text.split('\n')
# Filter out empty lines and take max 4 lines
final_poem = '\n'.join([line for line in poem_lines if line.strip()][:4])

print(prompt + final_poem)
```
