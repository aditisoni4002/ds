#Feature Engineering

1. What is a parameter?

- In machine learning, a parameter is a variable that is internal to the model and whose value is learned from the training data. Examples include the weights and biases in a neural network.

2. What is correlation? What does negative correlation mean?
- Correlation is a statistical measure that describes the extent to which two variables change together.

- A negative correlation means that as one variable increases, the other tends to decrease.

3. Define Machine Learning. What are the main components in Machine Learning?

- Machine learning is a field of artificial intelligence that focuses on developing algorithms that allow computers to learn from data without being explicitly programmed. The main components typically include:

- Data: The input information the model learns from.
- Model: The algorithm or structure that learns from the data.
- Objective Function (Loss Function): A function that measures how well the model is performing.

- Optimization Algorithm: A method used to adjust the model's parameters to minimize the loss function.

4. How does loss value help in determining whether the model is good or not?

- The loss value quantifies the error between the model's predictions and the actual values. A lower loss value generally indicates a better-performing model, as it means the model's predictions are closer to the true values.

5. What are continuous and categorical variables?

- Continuous variables can take any value within a given range (e.g., temperature, height, price).

- Categorical variables can only take a limited number of distinct values (e.g., color, gender, country).

6. How do we handle categorical variables in Machine Learning? What are the common techniques?

- Machine learning models typically require numerical input, so categorical variables need to be converted. Common techniques include: * One-Hot Encoding: Creates a new binary column for each unique category. * Label Encoding: Assigns a unique integer to each category. * Ordinal Encoding: Assigns integers based on the order of categories (if there is one).

7. What do you mean by training and testing a dataset?

- Training: The process of using a portion of the data (the training set) to teach the model to make predictions.

- Testing: The process of evaluating the trained model's performance on a separate, unseen portion of the data (the test set) to see how well it generalizes to new data.

8. What is sklearn.preprocessing?

- sklearn.preprocessing is a module in the scikit-learn library that provides functions and classes for data preprocessing, such as scaling, encoding categorical features, imputation, etc.

9. What is a Test set?

- A test set is a subset of the data that is held out and used to evaluate the performance of a machine learning model after it has been trained on the training set.

10. How do we split data for model fitting (training and testing) in Python?

- The train_test_split function from sklearn.model_selection is commonly used to split data into training and testing sets.

11. How do you approach a Machine Learning problem?

- A common approach includes:
- Understanding the problem: Define the objective and the data available.
- Data Collection: Gather relevant data.
- Exploratory Data Analysis (EDA): Understand the data's structure, identify patterns, and find insights.
- Data Preprocessing: Clean, transform, and prepare the data for modeling.
- Feature Engineering: Create new features from existing ones.
- Model Selection: Choose appropriate machine learning models.
- Model Training: Train the chosen models on the training data.
- Model Evaluation: Assess the models' performance using appropriate metrics on the test data.
- Model Tuning: Optimize model hyperparameters.
- Deployment: Integrate the best-performing model into a system.

16. What is causation? Explain difference between correlation and causation with an example.
- Causation means that one event is the direct result of another event. Correlation does not imply causation.
- Example: There is a positive correlation between ice cream sales and drowning incidents. However, ice cream sales do not cause drowning. Both are correlated because they both increase during hot summer months. The hot weather is the causal factor.
17. What is an Optimizer? What are different types of optimizers? Explain each with an example.
- An optimizer is an algorithm used to minimize the loss function by adjusting the model's parameters during training. Different types of optimizers include:
- Gradient Descent: A basic optimizer that updates parameters in the direction opposite to the gradient of the loss function.
- Stochastic Gradient Descent (SGD): Updates parameters using the gradient of a single training example or a small batch.
- Adam: An adaptive learning rate optimization algorithm that combines the ideas of momentum and RMSprop.

22. What is feature scaling? How does it help in Machine Learning?

- Feature scaling is a technique used to standardize or normalize the range of independent variables or features of the data. It helps in Machine Learning by:
- Preventing features with larger values from dominating those with smaller values, especially in algorithms sensitive to the scale of the data (e.g., support vector machines, k-nearest neighbors, gradient descent-based methods).
- Speeding up the convergence of optimization algorithms.
